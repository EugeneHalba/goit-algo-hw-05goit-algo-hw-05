```python
# USE REFERENCE

def fill_avg_salary(s, ref_df):
  avg = None
  try:
    sp = s.get('Посада')
    if sp:
      stat_sp = ref_df.loc[sp]
      avg=int((stat_sp["min"] + stat_sp["max"])/2)
  finally:
    s['avg']=avg
    return s


a=df.apply(fill_avg_salary, args=[spec_stat], axis=1)
a[a.columns[[0,1,2,19,20]]].head(2)
```
